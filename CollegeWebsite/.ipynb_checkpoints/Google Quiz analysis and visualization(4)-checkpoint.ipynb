```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

df = pd.read_csv("Question test -1.csv") # pass the google forms quiz responses csv file 
df

```python
df.info()
```

```python
df['Total score']
```

```python
df['Secured score'] = df['Total score'].str.replace(r'/.*', '').astype(float)
```

```python
df['Total score'] = df['Total score'].str.replace(r'.*/', '').astype(float)
```

```python
df.head()
```

```python
columns_needed = ['Roll no', 'Total score', 'Secured score']
for column in df.columns:
    d = re.match(r'.*\[Score\]', column)
    if d:
        columns_needed.append(d.group(0))
```

```python
df_needed = df[columns_needed][:]
df_needed
```

```python
columns_to_be_plotted = []
for column in columns_needed[3:]:
    total_score = df_needed[column].str.replace(r'.*/', '').astype(float)
    if any(total_score) > 0:
        columns_to_be_plotted.append(column)
        df_needed['Total ' + column] = df_needed[column].str.replace(r'.*/', '').astype(float)
        df_needed['Secured ' + column] = df_needed[column].str.replace(r'/.*', '').astype(float)
    df_needed = df_needed.drop(column, axis=1)
```

```python
df_needed.head()
```

```python
df_needed.info()
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
data = df_needed['Secured score'].value_counts()
x = data.index
y = data.values
ax.bar(x, y, width=0.4)
ax.set_title('Total points distribution', fontsize=22)
ax.set_xlabel("Points Scored", fontsize=18)
ax.set_ylabel("Number of students", fontsize=18)
ax.set_xlim(df['Secured score'].min() - 1, df['Secured score'].max() + 1)
ymin, ymax = plt.ylim()
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
ax.set_yticks(range(int(ymin), int(ymax) + 1));
ax.axhline(df_needed['Secured score'].mean(), ls='--', c="black");
```

```python
for column in columns_to_be_plotted:
        df_needed['Total ' + column] = df_needed['Total ' + column].fillna(df_needed['Total ' + column].mode()[0])
df_needed
```

```python
def plot_q(df, name):
    not_attempted = df['Secured '+name].isna().sum()
    attempted = len(df['Secured '+name]) - not_attempted
    data = df['Total '+name] == df['Secured '+name]
    d = {'Attempted': attempted, 
        'Not Attempted': not_attempted,
        'Correct Responses':data.value_counts().loc[True], 
         'Incorrect Responses': data.value_counts().loc[False] - not_attempted}
    temp = d.copy()
    temp['Question'] = name
    plot_bar_graph(list(d.keys()), list(d.values()), 'Q. '+name)
    return temp
```

```python
def plot_bar_graph(x, y, title):
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.set_title(title, fontsize=20)
    ax.bar(x, y, width=0.3, color=['#003f5c', '#ffa600'])
    ax.set_ylabel("Number of students", fontsize=15)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    ymin, ymax = plt.ylim()
    ax.set_yticks(range(int(ymin), int(ymax) + 1));
```

```python
q_info = []
for c in columns_to_be_plotted:
    q_info.append(plot_q(df_needed, c))
```

```python
q_df = pd.DataFrame(q_info).set_index('Question')
q_df # contains questions_info
```

```python
def stud_statistics(series, columns_to_be_plotted):
    d = {'Roll No': series['Roll no'],
         'Total Score': series['Total score'],
         'Secured Score': series['Secured score'],
        'Attempted': 0, 
    'Not Attempted': 0,
    'Correct Responses': 0, 
     'Incorrect Responses': 0}
    for col in columns_to_be_plotted:
        if pd.isna(series['Secured '+ col]):
            d['Not Attempted'] += 1
        else:
            d['Attempted'] += 1
            if series['Total ' + col] == series['Secured '+ col]:
                d['Correct Responses'] += 1
            else:
                d['Incorrect Responses'] += 1
    return d
```

```python
student_info = []
for index, row in df_needed.iterrows():
    student_info.append(stud_statistics(row, columns_to_be_plotted))
```

```python
stud_info = pd.DataFrame(student_info).set_index('Roll No')
```
